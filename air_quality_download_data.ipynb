```python
import pandas as pd

df['Dates'] = pd.to_datetime(df['ReadingDateTimeUTC']).dt.date
df['Time'] = pd.to_datetime(df['ReadingDateTimeUTC']).dt.time

df.drop(['City','ReadingDateTimeUTC'],axis=1)
```

```python
groupedbydevice = df.groupby(['DeviceId', 'Dates'])

groupedbydevice.first()
```

```python
# only keeping highest values recorded by each sensor 

device = df.sort_values(by=['DeviceId'])

df2 = device.sort_values(by=['PM25'], ascending=False)
robyn = df2.groupby(['DeviceId']).first()

robyn
```

```python
gfg_csv_data = robyn.to_csv(r'C:\Users\eugenie\Desktop\robyn_dataframe.csv')
```
